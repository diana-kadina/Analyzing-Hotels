### Looking for the Date

```python
df_date.head()
```

```python
df = pd.merge(df, df_date, left_on='check_in_date', right_on='date')
df.head()
```

```python
df.groupby('day_type')['occupancy_percentage'].mean().round(2).reset_index().sort_values(by='occupancy_percentage', ascending=False)
```

```python
df['month'] = df['date'].dt.strftime('%B')
df['year'] = df['date'].dt.year
df_new = df.groupby(['month', 'city'])['occupancy_percentage'].mean().round(2).reset_index()
df_new

```

```python
pivot_df = df_new.pivot(index='month', columns='city', values='occupancy_percentage')
pivot_df
```

```python
pivot_df.plot(kind='bar', figsize=(14, 6), color=colors)
plt.xlabel('Month')
plt.ylabel('Occupancy Percentage')
plt.title('Mean Occupancy Percentage by Month and City')
plt.tight_layout()
plt.show()
```

```python
df_bookings_all = pd.merge(df_hotels, df_bookings, on='property_id')
df_bookings_all.head()
```

```python
revenue_per_city = df_bookings_all.groupby('city')['revenue_realized'].sum().sort_values(ascending=False).reset_index()
revenue_per_city.plot(kind='bar', x='city', y='revenue_realized', figsize=(14, 8), color=colors)
plt.title('Revenue Realized per City')
plt.xlabel('City')
plt.ylabel('Realized Revenue')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
df_revenue_per_property = df_bookings_all.groupby('property_name')['revenue_realized'].sum().sort_values(ascending=False).reset_index()
df_revenue_per_property
```

```python
pie_colors=['darkred', 'coral', 'forestgreen', 'royalblue', 'lavender', 'blueviolet', 'plum']
df_revenue_per_property.set_index('property_name')['revenue_realized'].plot(
    kind='pie', colors=pie_colors, figsize=(6, 4), autopct='%1.1f%%'
)
plt.title('Revenue Distribution per Property')
plt.ylabel('')  
plt.tight_layout()
plt.show()
```

```python
df_ratings = df_bookings_all.groupby("property_name")["ratings_given"].mean().round(2).sort_values(ascending=False).reset_index()
df_ratings.plot(
    kind='bar', x='property_name', y='ratings_given', color=colors, figsize=(16, 8))
plt.title('Rating per Property')
plt.ylabel('')  
plt.tight_layout()
plt.show()
```
